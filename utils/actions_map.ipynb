```python
from itertools import product
```

```python
phi=[0,1,2,3,-3,-2,-1]
theta = [0,1,2,3,6,11,22,45,-45,-22,-11,-6,-3,-2,-1]
```

```python
actions = list(product(theta,phi))
```

```python
a_map = {}
for i in range(len(actions)):
    a_map[i] = actions[i]
```

```python
cont = 0
print('{',end="")
for i in range(len(a_map)):
    print("{} : ({},{}), ".format(i,a_map[i][0],a_map[i][1]),end="")
    if (i+1)%5 ==0:
        print("")
print('}')
```
